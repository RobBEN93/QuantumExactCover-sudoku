```python
from sudoku_nisq.data_generation import GenData, QuantumDataAnalysis
```

We generate random puzzles with $15-25$ missing cells.

```python
# Example use of QuantumDataAnalysis class
gen_data = GenData(grid_size=3, num_puzzles=10000)
data_df = gen_data.generate_data(num_cells_range=(15,25))
analysis = QuantumDataAnalysis(df=data_df)
```

```python
# Get descriptive statistics
stats = analysis.get_statistics()
print(stats)
```
