```python
from sudoku_nisq.q_sudoku import Sudoku
```

Generate and plot a random 4x4 puzzle with 4 missing cells:

```python
board = [[0,1,2,4],[2,4,0,0],[4,2,1,3],[1,0,4,2]]
sudoku = Sudoku(board=board,grid_size=2)
sudoku.plot()
```

```python
ibm_token = ""
sudoku.init_quantum(simple=True,pattern=False)
sudoku.quantum.init_ibm(ibm_token)
```

```python
print(sudoku.quantum.find_transp_resources(ibm_backend="ibm_brisbane",optimisation_levels=[0],flatten=1))
print(sudoku.quantum.find_transp_resources(ibm_backend="ibm_brisbane",optimisation_levels=[1]))
print(sudoku.quantum.find_transp_resources(ibm_backend="ibm_brisbane",optimisation_levels=[2]))
```
